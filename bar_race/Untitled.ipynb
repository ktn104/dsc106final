```python
import pandas as pd
import numpy as np
```

```python
players = pd.read_csv('Seasons_Stats.csv')
```

```python
df = players.groupby(['Year','Player']).sum().groupby('Player').cumsum()
```

```python
df = df.reset_index()
df
```

```python
sub = df[['Year','Player','PTS']]
```

```python
sub
```

```python
sub.pivot_table(values='PTS', index=sub.Player, columns='Year')
```

```python
test =  sub.reset_index().groupby(['Player', 'Year'])['PTS'].aggregate('first').unstack()
```

```python
test = test.fillna(axis=1, method='ffill')
```

```python
test = test.fillna(0)
```

```python
test.to_csv('done.csv')
```
